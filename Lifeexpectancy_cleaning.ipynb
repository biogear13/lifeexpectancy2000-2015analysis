# Data Cleaning

```python
lifeexpectancy.info()
```

- Change the column names to remove the spaces

```python
le2 = lifeexpectancy.rename({'Life expectancy ':'life_expectancy', 'Adult Mortality':'adult_mortality', 'infant deaths':'infant_deaths',
                             'percentage expenditure':'perc_expend', 'Hepatitis B':'hepa_b', ' BMI ':'BMI',
                             'under-five deaths ': 'underfive_deaths', 'Total expenditure': 'total_expenditure', 'Diphtheria ': 'diphtheria',
                             ' HIV/AIDS':'hiv', ' thinness  1-19 years':'thinness_10_19', ' thinness 5-9 years':'thinness_5_9',
                             'Income composition of resources':'income_comp'}, axis=1)
le2.info()
```

```python
#change year to text
le2['Year']=le2['Year'].astype(str)
```

```python
le2.shape
```

```python
# get the corrected country status
status = pd.read_csv('country status.csv')
status.head()
```

```python
# merge status table to le2
map_status = dict(status[['Country','Status_revised']].values)
le2['Status'] = le2['Country'].map(map_status)
```

Evaluate Population

```python
# What are the countries with missing population
le2[le2['Population'].isnull()]['Country'].unique()
```

Since there are significant number of items that were missing for population and gdp, I supplemented it from data from the world bank

```python
# get the population
population = pd.read_csv('population.csv')
population.head()
```

```python
# melt the population df
population_melted = pd.melt(population, id_vars=['Country Name'], value_vars=['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009',
                                                                              '2010','2011','2012','2013','2014','2015'], var_name='Year', value_name='population_wb')
population_melted
```

```python
#get the gdp
gdp = pd.read_csv('gdp.csv')
# melt the population df
gdp_melted = pd.melt(gdp, id_vars=['Country Name'], value_vars=['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009',
                                                                              '2010','2011','2012','2013','2014','2015'], var_name='Year', value_name='gdb_wb')
gdp_melted
```

```python
#join population and gdp
pop_gdp=population_melted.merge(gdp_melted, on=['Country Name','Year'], how='left')
pop_gdp.info()
```

```python
#join pop_gdp to le2
le3 = le2.merge(pop_gdp, left_on=['Country','Year'], right_on=['Country Name','Year'], how='left')
le3 = le3.drop(columns=['GDP','Population','Country Name'])
```

```python
le3.isnull().sum().divide(2928).multiply(100)
```

```python
le3[le3['gdb_wb'].isnull()][['Country','Year']]
```
